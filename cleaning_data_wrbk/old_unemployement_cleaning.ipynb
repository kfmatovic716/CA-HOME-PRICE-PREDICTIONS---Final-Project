```python
# Dependencies and Setup
import pandas as pd
import pathlib
import csv
import scipy.stats as st
import numpy as np
import calendar
from time import strptime
import datetime
```

```python
# Create paths for csv
CA_unemployment = "./unemployment_dataset/CA_unemployment_rate_data.txt"
```

```python
# Read csv
CA_unemployment_df = pd.read_csv(CA_unemployment)
pd.DataFrame(CA_unemployment_df)
```

```python
# Combine Year and Period columns to create Date columns
CA_unemployment_df['Date'] = CA_unemployment_df.Year.map(str) + " " + CA_unemployment_df.Period
CA_unemployment_df
```

```python
# Move the Date up to the front
date_CA_unemployment_df = CA_unemployment_df[ ['Date'] + [ col for col in CA_unemployment_df.columns if col != 'Date' ] ]
date_CA_unemployment_df
```

```python
# Convert Date column to YYYY-MM-DD
# Option 1: Use the date column and read the format and change to YYYY-MM. THEN add 31st (-31) at the end of all the dates
# Option 2: FIRST reformat all the Period column to MM (e.g. 01, 02, ...). THEN combine Year + Period together. THEN add 31st to all the dates. 

temp = datetime.datetime.strptime('2021 Jan', '%Y %m')
result = temp.strftime('%Y-%m')
```

```python
# Delete extraneous columns
delete_extraneous_columns = date_CA_unemployment_df.drop(columns= ["Adjusted", "Preliminary"])    
delete_extraneous_columns.head()
```

```python
# Remove the county from each name
delete_extraneous_columns["Area"] = delete_extraneous_columns["Area"].str.replace(" County", "")
delete_extraneous_columns
```
